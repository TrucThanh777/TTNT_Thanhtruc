```python
# --- Ki·ªÉm tra ng∆∞·ªùi th·∫Øng ---
def check_winner(board):
    win_states = [
        [board[0][0], board[0][1], board[0][2]],
        [board[1][0], board[1][1], board[1][2]],
        [board[2][0], board[2][1], board[2][2]],
        [board[0][0], board[1][0], board[2][0]],
        [board[0][1], board[1][1], board[2][1]],
        [board[0][2], board[1][2], board[2][2]],
        [board[0][0], board[1][1], board[2][2]],
        [board[0][2], board[1][1], board[2][0]],
    ]

    if ["X", "X", "X"] in win_states:
        return "X"
    if ["O", "O", "O"] in win_states:
        return "O"
    return None


# --- Ki·ªÉm tra c√≤n √¥ tr·ªëng kh√¥ng ---
def is_moves_left(board):
    for row in board:
        if "" in row:
            return True
    return False
```

```python
import math

def minimax(board, depth, is_max):
    winner = check_winner(board)

    # ƒêi·ªÉm cho c√°c tr·∫°ng th√°i k·∫øt th√∫c
    if winner == "O":
        return 10 - depth
    if winner == "X":
        return depth - 10
    if not is_moves_left(board):
        return 0

    # L∆∞·ª£t AI ‚Äî maximizing
    if is_max:
        best = -math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == "":
                    board[i][j] = "O"
                    best = max(best, minimax(board, depth+1, False))
                    board[i][j] = ""
        return best

    # L∆∞·ª£t ng∆∞·ªùi ch∆°i ‚Äî minimizing
    else:
        best = math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == "":
                    board[i][j] = "X"
                    best = min(best, minimax(board, depth+1, True))
                    board[i][j] = ""
        return best
```

```python
def find_best_move(board):
    best_val = -math.inf
    best_move = (-1, -1)

    for i in range(3):
        for j in range(3):
            if board[i][j] == "":
                board[i][j] = "O"  # AI th·ª≠ ƒë√°nh
                move_val = minimax(board, 0, False)
                board[i][j] = ""

                if move_val > best_val:
                    best_val = move_val
                    best_move = (i, j)

    return best_move
```

```python
def print_board(board):
    print("\n----- B√†n c·ªù -----")
    for row in board:
        print("|", " | ".join(c if c != "" else " " for c in row), "|")
    print("------------------\n")
```

```python
def play_game():
    board = [["", "", ""],
             ["", "", ""],
             ["", "", ""]]

    print("B·∫°n l√† X ‚Äî AI l√† O")
    print_board(board)

    while True:
        # --- Ng∆∞·ªùi ch∆°i ---
        row = int(input("H√†ng (0-2): "))
        col = int(input("C·ªôt (0-2): "))

        if board[row][col] != "":
            print("‚õî √î n√†y ƒë√£ ƒë√°nh r·ªìi! Ch·ªçn √¥ kh√°c.")
            continue

        board[row][col] = "X"
        print_board(board)

        if check_winner(board) == "X":
            print("üéâ B·∫°n th·∫Øng!")
            break
        if not is_moves_left(board):
            print("ü§ù H√≤a!")
            break

        # --- AI ƒë√°nh ---
        print("ü§ñ AI ƒëang suy nghƒ©...")
        ai_move = find_best_move(board)
        board[ai_move[0]][ai_move[1]] = "O"
        print_board(board)

        if check_winner(board) == "O":
            print("ü§ñ AI th·∫Øng!")
            break
        if not is_moves_left(board):
            print("ü§ù H√≤a!")
            break
```

```python
play_game()
```
