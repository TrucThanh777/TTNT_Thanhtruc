**Thuật toán **

huật toán K-means là một trong những thuật toán phân cụm (clustering) phổ biến nhất trong học máy không giám sát. Mục tiêu của thuật toán là chia tập dữ liệu thành K nhóm riêng biệt, sao cho các điểm trong cùng một nhóm giống nhau nhất có thể, và khác biệt tối đa so với các nhóm khác. Thuật toán hoạt động dựa trên ý tưởng tối ưu hóa khoảng cách giữa các điểm dữ liệu với tâm cụm (centroid).

K-means bắt đầu bằng việc khởi tạo K tâm cụm, thường được chọn ngẫu nhiên từ tập dữ liệu. Tuy nhiên, để cải thiện chất lượng phân cụm, kỹ thuật K-means++ thường được sử dụng để chọn các tâm khởi tạo sao cho chúng phân bố xa nhau, giúp thuật toán hội tụ nhanh hơn và tránh các cực trị cục bộ.

Sau khi có tâm cụm ban đầu, thuật toán thực hiện lặp lại hai bước cơ bản:

Bước 1: Gán dữ liệu vào cụm gần nhất. Tất cả các điểm dữ liệu được đo khoảng cách tới từng tâm cụm. Điểm nào gần tâm nào nhất sẽ thuộc về cụm đó. Thước đo phổ biến nhất là khoảng cách Euclid, nhưng trong nhiều trường hợp có thể sử dụng Manhattan, Cosine hoặc các hàm đo khác tùy bài toán.

Bước 2: Cập nhật lại tâm cụm. Sau khi tập dữ liệu được gán vào từng cụm, tâm cụm mới được tính lại bằng trung bình cộng của các điểm thuộc cụm đó. Tâm này đại diện cho vị trí trung tâm mới của nhóm.

Hai bước trên được lặp liên tục cho đến khi:

Tâm cụm không còn thay đổi đáng kể sau mỗi vòng lặp, hoặc

Đạt đến số vòng lặp tối đa mà thuật toán quy định.

Do đặc tính tối ưu cục bộ, K-means có thể hội tụ nhanh nhưng không đảm bảo tìm được nghiệm tối ưu toàn cục. Vì vậy, để đạt kết quả tốt hơn, người ta thường chạy K-means nhiều lần với các tâm khởi tạo khác nhau, sau đó chọn mô hình có tổng lỗi nhỏ nhất.

K-means hoạt động hiệu quả trên dữ liệu lớn nhờ độ phức tạp thấp và tính toán đơn giản. Tuy nhiên, thuật toán hoạt động tốt nhất khi dữ liệu có dạng hình cầu (cluster dạng tròn), kích thước cụm tương đối cân bằng, và không có quá nhiều nhiễu. Trong các bài toán thực tế, K-means đóng vai trò quan trọng trong phân tích dữ liệu, phát hiện cấu trúc ẩn, giảm chiều dữ liệu và làm bước tiền xử lý trước khi áp dụng các thuật toán học máy khác.

**Bài 1. Cài đặt lại thuật toán k-means.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo dữ liệu mẫu để thử nghiệm K-means
np.random.seed(42)
X1 = np.random.randn(100, 2) + np.array([2, 2])
X2 = np.random.randn(100, 2) + np.array([-2, -2])
X3 = np.random.randn(100, 2) + np.array([2, -2])
X = np.vstack((X1, X2, X3))
```

```python
# Hàm tính khoảng cách Euclid
def euclidean_distance(x, c):
    return np.sqrt(np.sum((x - c) ** 2))

# Hàm cập nhật tâm cụm
def update_centroid(points):
    return np.mean(points, axis=0)

# Gán điểm vào cụm gần nhất
def assign_cluster(x, centroids):
    distances = [euclidean_distance(x, c) for c in centroids]
    return np.argmin(distances)

# Thuật toán K-means chính
def kmeans(X, K, max_iters=100):
    # 1. Chọn ngẫu nhiên K tâm cụm
    np.random.seed(0)
    centroids = X[np.random.choice(len(X), K, replace=False)]

    for _ in range(max_iters):
        # 2. Gán điểm vào cụm
        clusters = [[] for _ in range(K)]
        for x in X:
            idx = assign_cluster(x, centroids)
            clusters[idx].append(x)

        old_centroids = np.copy(centroids)

        # 3. Cập nhật tâm cụm
        for j in range(K):
            if len(clusters[j]) > 0:
                centroids[j] = update_centroid(clusters[j])

        # 4. Kiểm tra hội tụ
        if np.all(old_centroids == centroids):
            break

    return centroids, clusters
```

```python
K = 3
centroids, clusters = kmeans(X, K)

# Vẽ đồ thị
colors = ['red', 'blue', 'green', 'purple', 'orange']

plt.figure(figsize=(7, 7))

for i in range(K):
    cluster = np.array(clusters[i])
    plt.scatter(cluster[:, 0], cluster[:, 1], s=30, color=colors[i], label=f'Cluster {i+1}')

# Vẽ tâm cụm
centroids = np.array(centroids)
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, color='black', marker='X', label='Centroids')

plt.legend()
plt.title("K-means Clustering Result")
plt.show()
```

**Cài đặt lại thuật toán k-NN với 2 cách đánh giá trọng số k.**

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b) ** 2))
```

```python
def knn_predict(X_train, y_train, x_test, k=3, weight="uniform"):
    # Tính khoảng cách từ x_test đến tất cả điểm trong X_train
    distances = [euclidean_distance(x_test, x) for x in X_train]

    # Lấy index của k hàng xóm gần nhất
    k_idx = np.argsort(distances)[:k]

    # Lấy nhãn tương ứng
    k_labels = y_train[k_idx]
    k_distances = np.array(distances)[k_idx]

    # ===== 1. TRỌNG SỐ ĐỀU (UNIFORM) =====
    if weight == "uniform":
        vote = Counter(k_labels)
        return vote.most_common(1)[0][0]

    # ===== 2. TRỌNG SỐ THEO KHOẢNG CÁCH (DISTANCE) =====
    elif weight == "distance":
        weights = 1 / (k_distances + 1e-9)  # tránh chia 0
        class_weights = {}

        for label, w in zip(k_labels, weights):
            class_weights[label] = class_weights.get(label, 0) + w

        # Chọn nhãn có tổng trọng số lớn nhất
        return max(class_weights, key=class_weights.get)

    else:
        raise ValueError("weight must be 'uniform' or 'distance'")
```

```python
# Tạo dữ liệu giả
np.random.seed(0)
X1 = np.random.randn(20, 2) + np.array([2, 2])
X2 = np.random.randn(20, 2) + np.array([-2, -2])

X_train = np.vstack((X1, X2))
y_train = np.array([0]*20 + [1]*20)

# Điểm cần dự đoán
x_test = np.array([1, 1])

# Dự đoán với 2 kiểu trọng số
pred_uniform = knn_predict(X_train, y_train, x_test, k=5, weight="uniform")
pred_distance = knn_predict(X_train, y_train, x_test, k=5, weight="distance")

print("Dự đoán (Uniform Weight):", pred_uniform)
print("Dự đoán (Distance Weight):", pred_distance)
```
