```python
# --- Ki·ªÉm tra ng∆∞·ªùi th·∫Øng ---
def check_winner(board):
    win_states = [
        [board[0][0], board[0][1], board[0][2]],
        [board[1][0], board[1][1], board[1][2]],
        [board[2][0], board[2][1], board[2][2]],
        [board[0][0], board[1][0], board[2][0]],
        [board[0][1], board[1][1], board[2][1]],
        [board[0][2], board[1][2], board[2][2]],
        [board[0][0], board[1][1], board[2][2]],
        [board[0][2], board[1][1], board[2][0]],
    ]

    if ["X", "X", "X"] in win_states:
        return "X"
    if ["O", "O", "O"] in win_states:
        return "O"
    return None


# --- Ki·ªÉm tra c√≤n √¥ tr·ªëng kh√¥ng ---
def is_moves_left(board):
    for row in board:
        if "" in row:
            return True
    return False
```

```python
import math

def minimax(board, depth, is_max):
    winner = check_winner(board)

    # ƒêi·ªÉm cho c√°c tr·∫°ng th√°i k·∫øt th√∫c
    if winner == "O":
        return 10 - depth
    if winner == "X":
        return depth - 10
    if not is_moves_left(board):
        return 0

    # L∆∞·ª£t AI ‚Äî maximizing
    if is_max:
        best = -math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == "":
                    board[i][j] = "O"
                    best = max(best, minimax(board, depth+1, False))
                    board[i][j] = ""
        return best

    # L∆∞·ª£t ng∆∞·ªùi ch∆°i ‚Äî minimizing
    else:
        best = math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == "":
                    board[i][j] = "X"
                    best = min(best, minimax(board, depth+1, True))
                    board[i][j] = ""
        return best
```

**C√†i ƒë·∫∑t thu·∫≠t to√°n c·∫Øt t·ªâa Alpha-beta v·ªõi ·ª©ng d·ª•ng v√†o b√†i to√°n TicTacToe v·ªõi ma

```python
import math

def check_winner(board):
    win_states = [
        [board[0][0], board[0][1], board[0][2]],
        [board[1][0], board[1][1], board[1][2]],
        [board[2][0], board[2][1], board[2][2]],
        [board[0][0], board[1][0], board[2][0]],
        [board[0][1], board[1][1], board[2][1]],
        [board[0][2], board[1][2], board[2][2]],
        [board[0][0], board[1][1], board[2][2]],
        [board[0][2], board[1][1], board[2][0]]
    ]

    if ["X", "X", "X"] in win_states:
        return "X"
    if ["O", "O", "O"] in win_states:
        return "O"
    return None


def is_moves_left(board):
    return any("" in row for row in board)
```

N·∫øu c√≥ ƒë∆∞·ªùng n√†o l√† ["X","X","X"] ‚Üí ng∆∞·ªùi ch∆°i X th·∫Øng

N·∫øu c√≥ ƒë∆∞·ªùng n√†o l√† ["O","O","O"] ‚Üí AI (O) th·∫Øng

Kh√¥ng c√≥ ai th·∫Øng ‚Üí tr·∫£ v·ªÅ None

üëâ T√≥m l·∫°i: H√†m n√†y ki·ªÉm tra xem X th·∫Øng, O th·∫Øng hay ch∆∞a ai th·∫Øng.

```python
def alphabeta(board, depth, alpha, beta, is_max):
    winner = check_winner(board)

    if winner == "O":
        return 10 - depth
    if winner == "X":
        return depth - 10
    if not is_moves_left(board):
        return 0

    if is_max:  # AI (O)
        best = -math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == "":
                    board[i][j] = "O"
                    value = alphabeta(board, depth + 1, alpha, beta, False)
                    board[i][j] = ""
                    best = max(best, value)
                    alpha = max(alpha, best)
                    if beta <= alpha:
                        return best
        return best

    else:  # Ng∆∞·ªùi ch∆°i (X)
        best = math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == "":
                    board[i][j] = "X"
                    value = alphabeta(board, depth + 1, alpha, beta, True)
                    board[i][j] = ""
                    best = min(best, value)
                    beta = min(beta, best)
                    if beta <= alpha:
                        return best
        return best
```

**C√†i ƒë·∫∑t thu·∫≠t to√°n minimax cho b√†i to√°n tictactoe v·ªõi ma tr·∫≠n 5x5, 10x10, NxN.**

```python
import math

# --- Ki·ªÉm tra ng∆∞·ªùi th·∫Øng tr√™n b·∫£ng NxN ---
def check_winner(board, K):
    N = len(board)

    # H√†ng ngang
    for i in range(N):
        for j in range(N - K + 1):
            line = board[i][j:j+K]
            if all(c == "X" for c in line):
                return "X"
            if all(c == "O" for c in line):
                return "O"

    # C·ªôt d·ªçc
    for i in range(N - K + 1):
        for j in range(N):
            line = [board[i+t][j] for t in range(K)]
            if all(c == "X" for c in line):
                return "X"
            if all(c == "O" for c in line):
                return "O"

    # Ch√©o ch√≠nh
    for i in range(N - K + 1):
        for j in range(N - K + 1):
            line = [board[i+t][j+t] for t in range(K)]
            if all(c == "X" for c in line):
                return "X"
            if all(c == "O" for c in line):
                return "O"

    # Ch√©o ph·ª•
    for i in range(N - K + 1):
        for j in range(K-1, N):
            line = [board[i+t][j-t] for t in range(K)]
            if all(c == "X" for c in line):
                return "X"
            if all(c == "O" for c in line):
                return "O"

    return None


# --- Ki·ªÉm tra c√≤n √¥ tr·ªëng kh√¥ng ---
def is_moves_left(board):
    return any("" in row for row in board)
```

Kh√¥ng c√≥ ai th·∫Øng ‚Üí tr·∫£ v·ªÅ None.

```python
def minimax(board, depth, is_max, K, max_depth):
    winner = check_winner(board, K)

    # Tr·∫°ng th√°i k·∫øt th√∫c
    if winner == "O":
        return 10 - depth
    if winner == "X":
        return depth - 10
    if not is_moves_left(board) or depth == max_depth:
        return 0

    N = len(board)

    # AI (O) - maximizing
    if is_max:
        best = -math.inf
        for i in range(N):
            for j in range(N):
                if board[i][j] == "":
                    board[i][j] = "O"
                    best = max(best, minimax(board, depth+1, False, K, max_depth))
                    board[i][j] = ""
        return best

    # Ng∆∞·ªùi ch∆°i (X) - minimizing
    else:
        best = math.inf
        for i in range(N):
            for j in range(N):
                if board[i][j] == "":
                    board[i][j] = "X"
                    best = min(best, minimax(board, depth+1, True, K, max_depth))
                    board[i][j] = ""
        return best
```

```python
def find_best_move(board, K, max_depth=3):
    N = len(board)
    best_val = -math.inf
    best_move = (-1, -1)

    for i in range(N):
        for j in range(N):
            if board[i][j] == "":
                board[i][j] = "O"
                move_val = minimax(board, 0, False, K, max_depth)
                board[i][j] = ""

                if move_val > best_val:
                    best_val = move_val
                    best_move = (i, j)

    return best_move
```

**C√†i ƒë·∫∑t thu·∫≠t to√°n c·∫Øt t·ªâa Alpha-beta cho b√†i to√°n tictactoe v·ªõi ma tr·∫≠n 5x5, 10x10,NxN.**


```python
import math

# Ki·ªÉm tra th·∫Øng: c·∫ßn K k√Ω t·ª± li√™n ti·∫øp ƒë·ªÉ th·∫Øng
def check_winner(board, K):
    N = len(board)

    # H√†ng ngang
    for i in range(N):
        for j in range(N - K + 1):
            line = board[i][j:j+K]
            if all(c == "X" for c in line): return "X"
            if all(c == "O" for c in line): return "O"

    # C·ªôt d·ªçc
    for i in range(N - K + 1):
        for j in range(N):
            line = [board[i+t][j] for t in range(K)]
            if all(c == "X" for c in line): return "X"
            if all(c == "O" for c in line): return "O"

    # Ch√©o ch√≠nh
    for i in range(N - K + 1):
        for j in range(N - K + 1):
            line = [board[i+t][j+t] for t in range(K)]
            if all(c == "X" for c in line): return "X"
            if all(c == "O" for c in line): return "O"

    # Ch√©o ph·ª•
    for i in range(N - K + 1):
        for j in range(K-1, N):
            line = [board[i+t][j-t] for t in range(K)]
            if all(c == "X" for c in line): return "X"
            if all(c == "O" for c in line): return "O"

    return None


# Ki·ªÉm tra c√≤n √¥ tr·ªëng kh√¥ng
def is_moves_left(board):
    return any("" in row for row in board)
```

N·∫øu t·∫•t c·∫£ h√†ng ƒë·ªÅu kh√¥ng c√≥ √¥ tr·ªëng ‚Üí tr·∫£ v·ªÅ False.

üëâ √ù nghƒ©a: ki·ªÉm tra game c√≤n ti·∫øp t·ª•c ƒë∆∞·ª£c hay ƒë√£ h√≤a.

```python
def alphabeta(board, depth, alpha, beta, is_max, K, max_depth):
    winner = check_winner(board, K)

    # Tr·∫°ng th√°i k·∫øt th√∫c
    if winner == "O": return 10 - depth
    if winner == "X": return depth - 10
    if not is_moves_left(board) or depth == max_depth: return 0

    N = len(board)

    # AI (O): Maximizing
    if is_max:
        best = -math.inf
        for i in range(N):
            for j in range(N):
                if board[i][j] == "":
                    board[i][j] = "O"
                    value = alphabeta(board, depth+1, alpha, beta, False, K, max_depth)
                    board[i][j] = ""
                    best = max(best, value)
                    alpha = max(alpha, best)

                    if beta <= alpha:  # C·∫Øt t·ªâa
                        return best
        return best

    # Ng∆∞·ªùi (X): Minimizing
    else:
        best = math.inf
        for i in range(N):
            for j in range(N):
                if board[i][j] == "":
                    board[i][j] = "X"
                    value = alphabeta(board, depth+1, alpha, beta, True, K, max_depth)
                    board[i][j] = ""
                    best = min(best, value)
                    beta = min(beta, best)

                    if beta <= alpha:  # C·∫Øt t·ªâa
                        return best
        return best
```

```python
def find_best_move_ab(board, K, max_depth=3):
    N = len(board)
    best_val = -math.inf
    best_move = (-1, -1)

    for i in range(N):
        for j in range(N):
            if board[i][j] == "":
                board[i][j] = "O"
                move_val = alphabeta(board, 0, -math.inf, math.inf, False, K, max_depth)
                board[i][j] = ""

                if move_val > best_val:
                    best_val = move_val
                    best_move = (i, j)

    return best_move
```
